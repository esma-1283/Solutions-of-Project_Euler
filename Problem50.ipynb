```python
Primes = []
for i in range(2,1000000):  
    is_prime = True
    for j in range(2, int(i ** 0.5) + 1):  
        if i % j == 0: 
            is_prime = False
            break 
    if is_prime:
        Primes.append(i)


sum=0
liste=[]
for i in Primes:
    sum+=i
    liste.append(sum)
print(liste)



solu=[]

for i in liste:
    if i in Primes:
            
        solu.append(i)

solu


```
